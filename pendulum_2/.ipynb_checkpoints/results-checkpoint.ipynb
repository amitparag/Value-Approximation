```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib 
matplotlib.use('GtkAgg')

import matplotlib.pyplot as plt
from tqdm import tqdm
from neural_net import feedForwardNet
from pendulum_utils import terminalPendulum, Solver, Datagen
import example_robot_data
import crocoddyl
from crocoddyl.utils.pendulum import CostModelDoublePendulum, ActuationModelDoublePendulum

#%matplotlib inline

net1 = torch.load('./nets/net1.pth')

#IREPA
net2 = torch.load('./nets/net2.pth')
net3 = torch.load('./nets/net3.pth')
net4 = torch.load('./nets/net4.pth')
net5 = torch.load('./nets/net5.pth')
net10 = torch.load('./nets/net10.pth')
net11 = torch.load('./nets/net11.pth')
net12 = torch.load('./nets/net12.pth')
net13 = torch.load('./nets/net13.pth')
net14 = torch.load('./nets/net14.pth')
net15 = torch.load('./nets/net15.pth')
net16 = torch.load('./nets/net16.pth')
net17 = torch.load('./nets/net17.pth')
net18 = torch.load('./nets/net18.pth')
net19 = torch.load('./nets/net19.pth')
net20 = torch.load('./nets/net20.pth')

grid_data = Datagen.samples(size=30,sampling='grid', th1_lims=[-0.99, 0.99], vel1_lims=[-0., 0])

v_croc = Datagen.statesValues(grid_data, as_tensor=False, full_traj=False)

p1 = net1(grid_data).detach().numpy().reshape(-1,1)

p5 = net5(grid_data).detach().numpy().reshape(-1,1)

p10 = net10(grid_data).detach().numpy().reshape(-1,1)

p15 = net15(grid_data).detach().numpy().reshape(-1,1)

p20 = net20(grid_data).detach().numpy().reshape(-1,1)

grids = grid_data.detach().numpy()
```

```python
plt.clf()

font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 10,
        }




fig, axs = plt.subplots(3, 2, figsize=(10, 12), sharex=True, sharey ='row')
fig.subplots_adjust(left=0.02, bottom=0.2, right=0.95, top=0.94, wspace=0.25)


# Plot prediction, cost
im1 = axs[0, 0].scatter(x = grids[:,0], y = grids[:,1], c= v_croc)
fig.colorbar(im1, ax=axs[0, 0])

im2 = axs[0, 1].scatter(x = grids[:,0], y = grids[:,1], c= p1)
fig.colorbar(im2, ax=axs[0, 1])

im3 = axs[1, 0].scatter(x = grids[:,0], y = grids[:,1], c= p5)
fig.colorbar(im3, ax=axs[1, 0])


im4 = axs[1, 1].scatter(x = grids[:,0], y = grids[:,1], c= p10)
fig.colorbar(im4, ax=axs[1, 1])


im5 = axs[2, 0].scatter(x = grids[:,0], y = grids[:,1], c= p15)
fig.colorbar(im5, ax=axs[2, 0])


im6 = axs[2, 1].scatter(x = grids[:,0], y = grids[:,1], c= p20)
fig.colorbar(im6, ax=axs[2, 1])


axs[0, 0].set_title("Value crocoddyl", font)
axs[0, 1].set_title("Irepa #1", font)
axs[1, 0].set_title("Irepa #5", font)
axs[1, 1].set_title("Irepa #10", font)
axs[2, 0].set_title("Irepa #15", font)
axs[2, 1].set_title("Irepa #20", font)

axs[0,0].set_xlim(-1, 1)
axs[0,0].set_ylim(-1, 1)

axs[0,0].set_xlabel("Theta 1", font)
axs[0,0].set_ylabel("Theta 2", font)
```
