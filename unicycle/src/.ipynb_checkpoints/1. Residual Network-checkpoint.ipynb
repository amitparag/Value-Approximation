```python
import numpy as np
import torch
import crocoddyl
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline


from terminal_unicycle import ResidualUnicycle
from dataset import Datagen
from residual_network import ResidualNet
from solver import solve_problem

res_net = torch.load("../networks/residual.pth")
```

```python


# Generate 100 starting positions from the circumference of circles of radius of 1.
test_data   = Datagen.circular_data(r=[2], n=[100])

xs    = []
value = []
iters = []
```

```python

def solutions(initial_configuration, precision=1e-6, maxiters = 1000):

    # Running model
    model  = crocoddyl.ActionModelUnicycle()
    model.costWeights = np.matrix([1., 1.]).T 

    # Terminal model, with residual network.
    model2 = ResidualUnicycle(res_net, use_gauss_approx=False)


    # Define the problem
    problem = crocoddyl.ShootingProblem(initial_configuration.T, [ model ] * 30, model2)
    ddp         = crocoddyl.SolverDDP(problem)

    # Set the precision to given value
    ddp.th_stop = precision

    #Solve the problem
    ddp.solve([], [], maxiters)

    return ddp
```

```python
for starting_position in test_data:
    ddp = solutions(initial_configuration = starting_position, precision = 1e-6, maxiters = 1000)
        
    xs.append(np.array(ddp.xs))
    value.append(ddp.cost)
    iters.append(ddp.iter)
```

```python



norm = mpl.colors.Normalize(vmin=min(value), vmax=max(value))
cmap = mpl.cm.ScalarMappable(norm = norm, cmap=mpl.cm.plasma)
cmap.set_array([])

for key, trajectory in zip(value, xs):
    plt.scatter(trajectory[:, 0], trajectory[:, 1], 
                    marker = '',
                    zorder=2, 
                    s=50,
                    linewidths=0.2,
                    alpha=.8, 
                    cmap = cmap )
    plt.plot(trajectory[:, 0], trajectory[:, 1], c=cmap.to_rgba(key))
plt.colorbar(cmap).set_label("Value", labelpad=2, size=15)
```

```python
plt.clf()
norm = mpl.colors.Normalize(vmin=min(iters), vmax=max(iters))
cmap = mpl.cm.ScalarMappable(norm = norm, cmap=mpl.cm.plasma)
cmap.set_array([])

for key, trajectory in zip(iters, xs):
    plt.scatter(trajectory[:, 0], trajectory[:, 1], 
                    marker = '',
                    zorder=2, 
                    s=50,
                    linewidths=0.2,
                    alpha=.8, 
                    cmap = cmap )
    plt.plot(trajectory[:, 0], trajectory[:, 1], c=cmap.to_rgba(key))
plt.colorbar(cmap).set_label("Iters", labelpad=2, size=15)
```
