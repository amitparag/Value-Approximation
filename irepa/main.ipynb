```python
import crocoddyl
from helper import torch
from tqdm import tqdm
import time
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from helper import Datagen
from helper import PolicyNetwork
from helper import ValueNetwork
import numpy as np
torch.set_default_dtype(torch.double)

if torch.cuda.is_available():
    torch.cuda.empty_cache()

```

```python
#...... Global Hyperparameters
TRAJECTORIES     = 500
HORIZON          = 30
PRECISION        = 1e-9
MAXITERS         = 1000

#...... Hyperparameters for value network

V_BATCHSIZE        = 32
V_fc1_dims         = 32
V_fc2_dims         = 32
V_fc3_dims         = 32
V_lr               = 1e-3
V_EPOCHS           = 1000
V_HORIZON          = 100
V_DECAY            = 0
V_activation       = torch.nn.Tanh()
V_DEVICE           = 'cpu'


#...... Hyperparameters for policy network

P_BATCHSIZE        = 64
P_fc1_dims         = 256
P_fc2_dims         = 256
P_fc3_dims         = 20
P_lr               = 1e-3
P_EPOCHS           = 1000
P_HORIZON          = 100
P_DECAY            = 0
P_activation       = torch.nn.Tanh()
P_DEVICE           = 'cpu'
```

```python
# Instantiate an empty value net and policy net

vnet = ValueNetwork(fc1_dims=V_fc1_dims, fc2_dims=V_fc2_dims, fc3_dims=V_fc3_dims, activation=V_activation, device=V_DEVICE)
pnet = PolicyNetwork(fc1_dims=P_fc1_dims, fc2_dims=P_fc2_dims, fc3_dims=P_fc3_dims, activation=P_activation, device=P_DEVICE)
```

```python
#...... Generate Dataset

starting_points         = Datagen.griddedData(n_points=TRAJECTORIES)

v_xtrain, v_ytrain = [], []
p_xtrain, p_ytrain = [], []

for starting_point in starting_points:
    model               = crocoddyl.ActionModelUnicycle()
    model.costWeights   = np.array([1.,1.]).T
    problem             = crocoddyl.ShootingProblem(starting_point.T, [model]*HORIZON, model)
    ddp                 = crocoddyl.SolverDDP(problem)
    ddp.th_stop         = PRECISION
    ddp.solve([], [], MAXITERS)
    # Dataset for value
    xs = np.array(ddp.xs).tolist()
    for node in xs:
        v_xtrain.append(node)

    cost = []

    for d in ddp.problem.runningDatas:
        cost.append(d.cost)
    cost.append(ddp.problem.terminalData.cost)

    for i, _ in enumerate(cost):
        cost[i] = sum(cost[i:])

    # Append costs in cost to values
    for c in cost:
        v_ytrain.append(c)

    # Dataset for policy
    for _ in ddp.xs[:-1]:
        p_xtrain.append(_)
    for _ in ddp.xs[1:]:
        p_ytrain.append(_)

v_xtrain, v_ytrain = np.array(v_xtrain), np.array(v_ytrain).reshape(-1, 1)
p_xtrain, p_ytrain = np.array(p_xtrain), np.array(p_ytrain)
```

```python
v_xtest  = torch.Tensor(v_xtrain[0:100,:]).to(V_DEVICE)
v_ytest  = torch.Tensor(v_ytrain[0:100,:]).to(V_DEVICE)

p_xtest  = torch.Tensor(p_xtrain[0:100,:]).to(P_DEVICE)
p_ytest  = torch.Tensor(p_ytrain[0:100,:]).to(P_DEVICE)

v_xtrain = torch.Tensor(v_xtrain[100:,:])
v_ytrain = torch.Tensor(v_ytrain[100:,:])

p_xtrain = torch.Tensor(p_xtrain[100:,:])
p_ytrain = torch.Tensor(p_ytrain[100:,:])
```

```python
# Convert to torch dataloader
v_dataset = torch.utils.data.TensorDataset(v_xtrain, v_ytrain)
v_dataloader = torch.utils.data.DataLoader(v_dataset, batch_size = V_BATCHSIZE, shuffle=True)

#......  CRITERIA
criterionv1 = torch.nn.MSELoss(reduction='sum')
criterionv2 = torch.nn.L1Loss(reduction='sum')


#.......  OPTIMIZER
#vopt = torch.optim.ASGD(vnet.parameters(), lr = V_lr, lambd=0.0001, alpha=0.75, t0=1000000.0, weight_decay= V_DECAY)
#vopt = torch.optim.SGD(vnet.parameters(), lr = V_lr, momentum=0.9, weight_decay=V_DECAY, nesterov=True)
vopt = torch.optim.Adam(vnet.parameters(), lr = V_lr, betas=[0.5, 0.9], weight_decay=V_DECAY)

for epoch in range(V_EPOCHS):
    for data, target in v_dataloader:
        vnet.train()
        vopt.zero_grad()

        data        = data.to(V_DEVICE)
        target      = target.to(V_DEVICE)
        
        output      = vnet(data)
        
        loss        = torch.sqrt(criterionv1(output, target)) + 0.01*criterionv2(output, target)
        print(loss)
        loss.backward()
        vopt.step()

    prediction  = vnet(v_xtest)
    mse         = torch.mean((prediction-v_ytest)**2)
    mae         = torch.mean(torch.abs(prediction-v_ytest))
    print(f"Epoch {epoch+1} :: mse = {mse}, mae = {mae}")
```

```python
accuracy = 0
for _ in range(100):
    x  = [np.random.uniform(-1.5,1.5), np.random.uniform(-1.5,1.5), np.random.uniform(-np.pi/4, np.pi/4)]
    np.round_(x, 4)

    x1 = torch.Tensor(x).to(V_DEVICE)
    cost = vnet(x1).item()

    x0 = np.array(x)
    model               = crocoddyl.ActionModelUnicycle()
    model.costWeights   = np.array([1.,1.]).T
    problem             = crocoddyl.ShootingProblem(x0.T, [model]*HORIZON, model)
    ddp                 = crocoddyl.SolverDDP(problem)
    ddp.solve([], [], MAXITERS)
    if np.abs(ddp.cost - cost) <= 2.:
        accuracy += 1
print(accuracy/100)
```

```python
p_dataset = torch.utils.data.TensorDataset(p_xtrain, p_ytrain)
p_dataloader = torch.utils.data.DataLoader(p_dataset, batch_size = P_BATCHSIZE, shuffle=True)

#......  CRITERIA
criterionp1 = torch.nn.MSELoss(reduction='sum')
criterionp2 = torch.nn.L1Loss(reduction='sum')


#.......  OPTIMIZER
#vopt = torch.optim.ASGD(vnet.parameters(), lr = V_lr, lambd=0.0001, alpha=0.75, t0=1000000.0, weight_decay= V_DECAY)
#vopt = torch.optim.SGD(vnet.parameters(), lr = V_lr, momentum=0.9, weight_decay=V_DECAY, nesterov=True)
popt = torch.optim.Adam(pnet.parameters(), lr = P_lr, betas=[0.5, 0.9], weight_decay=P_DECAY)

for epoch in range(P_EPOCHS):
    for data, target in p_dataloader:
        pnet.train()
        popt.zero_grad()

        data        = data.to(P_DEVICE)
        target      = target.to(P_DEVICE)
        
        output      = pnet(data)
        
        loss        = criterionp1(output, target) 
        
        loss.backward()
        popt.step()

    prediction  = pnet(p_xtest)
    mse         = torch.mean((prediction-p_ytest)**2)
    mae         = torch.mean(torch.abs(prediction-p_ytest))
    print(f"Epoch {epoch+1} :: mse = {mse}, mae = {mae}")
```
